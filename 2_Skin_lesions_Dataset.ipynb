```python
orig_folder= 'C:/Users/kravc/Final_Project/Data/Skin_classifier/lesion_images/'
train_folder = 'C:/Users/kravc/Final_Project/Data/Skin_classifier/skin_2cl/training/'
val_folder = 'C:/Users/kravc/Final_Project/Data/Skin_classifier/skin_2cl/validation/'
test_folder = 'C:/Users/kravc/Final_Project/Data/Skin_classifier/skin_2cl/test/'
train_0 = 'C:/Users/kravc/Final_Project/Data/Skin_classifier/skin_2cl/training/0_Non_Cancerous'
train_1 = 'C:/Users/kravc/Final_Project/Data/Skin_classifier/skin_2cl/training/1_Cancerous'
val_0 = 'C:/Users/kravc/Final_Project/Data/Skin_classifier/skin_2cl/validation/0_Non_Cancerous'
val_1 = 'C:/Users/kravc/Final_Project/Data/Skin_classifier/skin_2cl/validation/1_Cancerous'
test_0 = 'C:/Users/kravc/Final_Project/Data/Skin_classifier/skin_2cl/test/0_Non_Cancerous'
test_1 = 'C:/Users/kravc/Final_Project/Data/Skin_classifier/skin_2cl/test/1_Cancerous'
```

```python
#Copy images into respective folders to use with Keras flow_from_directory ImageDataGenerator

image_ids = df['image_id'].tolist()
subsets = df['set'].tolist()
labels = df['cancerous'].tolist()

for index, image in enumerate(image_ids):
    image = image + '.jpg'
    if subsets[index] == 'train':
        if labels[index] == 0:
            copyfile(os.path.join(orig_folder, image), os.path.join(train_0, image))
        elif labels[index] == 1:
            copyfile(os.path.join(orig_folder, image), os.path.join(train_1, image))
    elif subsets[index] == 'val':
        if labels[index] == 0:
            copyfile(os.path.join(orig_folder, image), os.path.join(val_0, image))
        elif labels[index] == 1:
            copyfile(os.path.join(orig_folder, image), os.path.join(val_1, image))
    elif subsets[index] == 'test':
        if labels[index] == 0:
            copyfile(os.path.join(orig_folder, image), os.path.join(test_0, image))
        elif labels[index] == 1:
            copyfile(os.path.join(orig_folder, image), os.path.join(test_1, image))        
```
